```python
########## DCODE(전입_지역코드) 생성 ##########
```

```python
df1['DCODE']=df1['전입행정_시도'].map(str)+df1['전입행정_시군구'].map(str)
df2['DCODE']=df2['전입행정_시도'].map(str)+df2['전입행정_시군구'].map(str)
```

```python
########## DCODEdummy(전입에 대한 더미함수) 생성 ##########
```

```python
# 전입에 대한 dummy 함수 생성, 해당 지역에 포함되면 1, 아니면 0으로 저장
dummy1=[]; dummy2=[]
values=[110, 111, 113, 115, 117, 280, 281, 283, 285, 287, 780, 
        130, 131, 132, 133, 134, 135, 460, 461, 462, 463, 465, 850]

for row in df1['전입행정_시군구']:
    if row in values:
        dummy1.append('1')
    else:
        dummy1.append('0')
        
for row in df2['전입행정_시군구']:
    if row in values:
        dummy2.append('1')
    else:
        dummy2.append('0')
        
df1['DCODEdummy']=dummy1; df2['DCODEdummy']=dummy2
```

```python
########## GID값이 있는 시트 불러오고 오류값으로 표기된 셀을 0.0 값으로 변경 ##########
```

```python
df_p=pd.read_excel('./files/codebook_final.xlsx', sheet_name='지역코드표', usecols=[0, 3])
df_p.head(128)
```

```python
df_p['GID+행정구+출장소']=df_p['GID+행정구+출장소'].fillna(0)
df_p.head(128)
```

```python
########## DGID 생성 ##########
```

```python
df_p['RCODE'].astype(int); df_p['GID+행정구+출장소'].astype(int)
_RCODE=[]; _GID=[]
for i in range(473):
    _RCODE.append(df_p['RCODE'][i])
    _GID.append(df_p['GID+행정구+출장소'][i])
    
DGID1=[]; DGID2=[]
for value in df1['DCODE']:
    index=_RCODE.index(int(value))
    DGID1.append(_GID[index])
    
for value in df2['DCODE']:
    index=_RCODE.index(int(value))
    DGID2.append(_GID[index])
    
df1['DGID']=DGID1; df2['DGID']=DGID2
```

```python
########## OCODE(전출_지역코드) 생성 ##########
```

```python
# 시도, 시군구의 값을 문자열로 바꾸고, 결합한 값은 문자열 타입으로 OCODE(전출_지역코드)에 저장
df1['OCODE']=df1['전출행정_시도'].map(str)+df1['전출행정_시군구'].map(str)
df2['OCODE']=df2['전출행정_시도'].map(str)+df2['전출행정_시군구'].map(str)
```

```python
########## OCODEdummy(전출_지역코드) 생성 ##########
```

```python
# 전출에 대한 dummy 함수 생성, 해당 지역에 포함되면 1, 아니면 0으로 저장
dummy3=[]; dummy4=[]
_values=[41110, 41111, 41113, 41115, 41117, 41280, 41281, 41283, 41285, 41287, 41780, 
       41130, 41131, 41132, 41133, 41134, 41135, 41460, 41461, 41462, 41463, 41465, 41850]

for row in df1['OCODE']:
    if int(row) in _values:
        dummy3.append('1')
    else:
        dummy3.append('0')
        
for row in df2['OCODE']:
    if int(row) in _values:
        dummy4.append('1')
    else:
        dummy4.append('0')
        
df1['OCODEdummy']=dummy3; df2['OCODEdummy']=dummy4
```
